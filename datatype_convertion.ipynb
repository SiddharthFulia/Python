```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("test.csv",header=0)
```

```python
#replace a certain value with second value of the argument
df['Product_Category_3'].replace(np.nan,0,inplace=True)
df['Product_Category_2'].replace(np.nan,0,inplace=True)
df['Product_Category_1'].replace(np.nan,0,inplace=True)
df['Stay_In_Current_City_Years'].replace('4+','4',inplace=True)
df.head()
```

```python
df["Age"].replace(['46-50', '26-35', '36-45', '18-25', '51-55', '55+', '0-17'],['50', '35', '45', '25', '55', '55', '17'],inplace=True)
```

```python
#Age column is not being shown because age is in form of object/string
df.describe()
```
